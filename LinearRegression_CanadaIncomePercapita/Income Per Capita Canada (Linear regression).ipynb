```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Read Data

data = pd.read_csv('IncomePercapita_Source.csv')
```

```python
#Verify Data import

data.info()
```

```python
#Visualize data distibution

sns.regplot(x='Year',y='Income',data=data)
```

```python
# regression
# create the instance
lri = LinearRegression()
```

```python
# Train model
lri.fit(data[['Year']].values, data.Income)
```

```python
lri.predict([[2021]])
```
